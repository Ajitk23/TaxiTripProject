## Database Creation

## Data Reading, Writing and Creating delta tables

## Trip Zone

```python
df_zone = spark.read.format('parquet')\
          .option('inferSchema', True)\
          .option('header', True)\
          .load('abfss://silver@nyccabstorage.dfs.core.windows.net/trip_zone')        
```

```python
df_zone.display()
```

```python
df_zone.write.format('delta')\
      .mode('append')\
      .option('path', 'abfss://gold@nyccabstorage.dfs.core.windows.net/trip_zone')\
      .saveAsTable('gold.trip_zone')        
```

## Trip Type

```python
df_type = spark.read.format('parquet')\
          .option('inferSchema', True)\
          .option('header', True)\
          .load('abfss://silver@nyccabstorage.dfs.core.windows.net/trip_type')
df_type.display()                  
```

```python
df_type.write.format('delta')\
       .mode('append')\
       .option('path', 'abfss://gold@nyccabstorage.dfs.core.windows.net/trip_type')\
       .saveAsTable('gold.trip_type')        
```

## Trips Data

```python
df_trips2023 = spark.read.format('parquet')\
               .option('inferSchema', True)\
               .option('header', True)\
               .load('abfss://silver@nyccabstorage.dfs.core.windows.net/tripdata2023')    
```
